#### Q3. What is a box tuple, and how does it work?

+ Tuple box is a tuple of four integer coordinates that represent a rectangular region in an image. e.g (Left,Top,Right,Bottom)
+ Left: The x-coordinate of the left most edge of the box.
+ Top: The y-coordinate of the top edge of the box.
+ Right: The x-coordinate of one pixel to the right of the rightmost edge of the box. This integer must be greater than the the top integer.
+ Bottom: The y-coordinate of the one pixel lower than the bottom edge of the box. This integer must be greater than the top integer.
